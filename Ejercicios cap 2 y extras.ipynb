```
#read the file into a Spark DataFrame
mnm_df = spark.read\
              .option("header", "true")\
              .option("inferSchema", "true")\
              .csv("/FileStore/tables/mnm_dataset.csv")
mnm_df.show(n=5, truncate=False)
```

```
# 1.Select from the DataFrame the fields "State", "Color", and "Count"
# 2. Since we want to group each state and its M&M color count, we use groupBy()
# 3. Aggregate counts of all colors and groupBy() State and Color
# 4 orderBy() in descending order

from pyspark.sql import functions as F
count_mnm_df = mnm_df.select("State", "Color", "Count")\
                     .groupBy("State", "Color")\
                     .agg(F.count("Count").alias("Total"))\
                     .orderBy(F.col("Total").desc())
```

```
# Show the resulting aggregations for all the states and colors;
# a total count of each color per state.
# Note show() is an action, which will trigger the above
# query to be executed

count_mnm_df.show(n=60, truncate = False)
print("Total Rows = %d" % (count_mnm_df.count()))
```

```
# While the above code aggregated and counted for all 
# the states, what if we just want to see the data for 
# a single state, e.g., CA? 
# 1. Select from all rows in the DataFrame
# 2. Filter only CA state
# 3. groupBy() State and Color as we did above
# 4. Aggregate the counts for each color
# 5. orderBy() in descending order 
# Find the aggregate count for California by filtering

count_mnm_df_CA = mnm_df.select("State", "Color", "Count")\
                        .filter((F.col("State") == "CA"))\
                        .groupBy("State", "Color")\
                        .agg(F.count("Count").alias("Total"))\
                        .orderBy(F.col("Total").desc())
    
count_mnm_df_CA.show()
```

```
# Otras operaciones de agregación como el Max con otro tipo de ordenamiento
max_mnm_df = mnm_df.select("State", "Color", "Count")\
                   .groupBy("Color")\
                   .agg(F.max("Count").alias("Maximo"))\
                   .orderBy(F.col("Maximo").desc())
max_mnm_df.show()


```

```
# hacer un ejercicio como el “where” de CA que aparece en el libro pero indicando más opciones de estados (p.e. NV, TX, CA, CO)
mnm_df_varios = mnm_df.select("State", "Color", "Count")\
                      .filter((F.col("State") == "NV") | (F.col("State") == "TX") | (F.col("State") == "CA") | (F.col("State") == "CO"))\
                      .groupBy("State", "Color")\
                      .agg(F.count("Count").alias("Total"))\
                      .orderBy(F.col("Total").desc())
                      
mnm_df_varios.show()
```
